# 2. The Question: Decision, Action, Cost of a Wrong Call

### Research Question

Which webpages should be reviewed first for content refresh based on their search performance and content signals?

### Decision

Should this webpage be reviewed for a content refresh?

### Unit of Analysis

One webpage.

### Output

A ranked refresh score showing which pages should be reviewed first.

### Action

The SEO team can:
- Refresh outdated content
- Improve keywords
- Expand thin content
- Improve CTR
- Monitor high-performing pages

### Cost of a Wrong Recommendation

**False Positive:** A good page is reviewed unnecessarily, wasting time.

**False Negative:** A page needing improvement is ignored, which may result in continued traffic loss.

Large websites contain thousands of pages. Machine learning can help prioritize review candidates using observable data, making the review process more efficient while keeping humans responsible for the final decisions.
